```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
data16 = np.load('data/Ly_16_it_100_da_0.npz')
data32 = np.load('data/Ly_32_it_100_da_0.npz')
data64 = np.load('data/Ly_64_it_100_da_0.npz')
data128 = np.load('data/Ly_128_it_100_da_0.npz')
data256 = np.load('data/Ly_256_it_100_da_0.npz')
data512 = np.load('data/Ly_512_it_100_da_0.npz')
data1024 = np.load('data/Ly_1024_it_10_da_0.npz')
```

```python
plt.figure()
plt.loglog(data16['t'],data16['u_sigma1'],label = 'Ly = 16')
plt.loglog(data32['t'],data32['u_sigma1'],label = 'Ly = 32')
plt.loglog(data64['t'],data64['u_sigma1'],label = 'Ly = 64')
plt.loglog(data128['t'],data128['u_sigma1'],label = 'Ly = 128')
plt.loglog(data256['t'],data256['u_sigma1'],label = 'Ly = 256')
plt.loglog(data512['t'],data512['u_sigma1'],label = 'Ly = 512')
plt.loglog(data1024['t'],data1024['u_sigma1'],label = 'Ly = 1024')
#plt.loglog(data16['t'],data16['u_sigma'],'--',label = 'Ly = 16')
#plt.loglog(data32['t'],data32['u_sigma'],'--',label = 'Ly = 32')
#plt.loglog(data64['t'],data64['u_sigma'],'--',label = 'Ly = 64')
#plt.loglog(data128['t'],data128['u_sigma'],'--',label = 'Ly = 128')
#plt.loglog(data256['t'],data256['u_sigma'],'--',label = 'Ly = 256')
#plt.loglog(data512['t'],data512['u_sigma'],'--',label = 'Ly = 512')
#plt.loglog(data1024['t'],data1024['u_sigma'],'--',label = 'Ly = 1024')

plt.xlabel('t')
plt.ylabel(r'$\omega$')
plt.legend()
plt.show()
```

```python
L = np.array([16,32,64,128,256,512,1024])
w_L = np.array([data['u_sigma'][-100:].mean() for data in [data16,data32,data64,data128,data256,data512,data1024]])
w_L1 = np.array([data['u_sigma1'][-100:].mean() for data in [data16,data32,data64,data128,data256,data512,data1024]])
```

```python
print(w_L)
print(w_L1)
```

```python
plt.figure()

zeta = .35
wi = w_L[0]*.8
wi1 = w_L1[0]*.1
plt.loglog(L,w_L**2-wi**2,'.',label = '$\omega^2$')
plt.loglog(L,w_L1**2-wi1**2,'.',label = '$\omega_1^2$')
plt.loglog(L,2e-1*L**(2*zeta),'--')
#plt.xlim(100,1100)

plt.legend()
plt.show()
```
